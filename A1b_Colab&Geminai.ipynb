```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Create an empty dictionary to store the rolls
rolls = {}

# Loop through the number of sides
for n in range(2, 11):
    # Roll the dice 100 times and store the results in a list
    rolls[n] = [random.randint(1, n) for _ in range(100)]

# Create a pandas DataFrame from the dictionary
df = pd.DataFrame(rolls)

# Count the number of times that 1 shows up for each die
ones_count = df.apply(lambda x: x.value_counts().get(1, 0))

# Create a bar plot of the number of times that 1 shows up
plt.bar(ones_count.index, ones_count.values)
plt.xlabel('Number of Sides')
plt.ylabel('Number of Times 1 Shows Up')
plt.title('Number of Times 1 Shows Up on Different Dice')
plt.show()
```

```python
# Create an empty dictionary to store the rolls
rolls = {}

# Loop through the number of sides
for n in range(2, 11):
    # Roll the dice 10,000 times and store the results in a list
    rolls[n] = [random.randint(1, n) for _ in range(10000)]

# Create a pandas DataFrame from the dictionary
df = pd.DataFrame(rolls)

# Find the most frequent value and its frequency for each die
most_frequent = df.apply(lambda x: x.value_counts().index[0])
frequency = df.apply(lambda x: x.value_counts().iloc[0])

# Print the results
for n in range(2, 11):
    print(f"For the {n}-sided die, the most frequent value is {most_frequent[n]} with a frequency of {frequency[n]}")
```
